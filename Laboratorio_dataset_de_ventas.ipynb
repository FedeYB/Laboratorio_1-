```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import glob
ruta_drive = '/content/drive/MyDrive/UPSO/UPSO/Segundo año/Segundo cuatrimestre/Elemento de aprendizaje de maquina y big data/Clase 3/Dataset de ventas2'
```

##**ENERO**

```python
# Carga y lectura del archivo CSV
enero = f'{ruta_drive}/Dataset_de_ventas_Enero.csv'

enero_df = pd.read_csv(enero)

# Limpieza y formateo de datos

enero_df = enero_df.dropna()

enero_df['Cantidad Pedida'] = pd.to_numeric(enero_df['Cantidad Pedida'], errors='coerce')

enero_df['Precio Unitario'] = pd.to_numeric(enero_df['Precio Unitario'], errors='coerce')

enero_df['Fecha de Pedido'] = pd.to_datetime(enero_df['Fecha de Pedido'], errors='coerce')

enero_df
```

```python
# Ventas totales por producto

ventas_por_producto_enero = enero_df.groupby('Producto')['Cantidad Pedida'].sum().reset_index()

ventas_por_producto_enero
```

```python
# Producto más vendido

producto_mas_vendido_enero = ventas_por_producto_enero.sort_values(by='Cantidad Pedida', ascending=False).iloc[0]

producto_mas_vendido_enero
```

```python
# Precio por unidad

precios_unitarios_enero = enero_df['Precio Unitario']
```

```python
# Crear un gráfico interactivo que muestre ventas totales como barras y precios unitarios como líneas
fig = go.Figure()

# Agregar la serie de barras para las ventas totales
fig.add_trace(go.Bar(x=ventas_por_producto_enero['Producto'], y=ventas_por_producto_enero['Cantidad Pedida'], name='Ventas Totales', yaxis='y'))

# Agregar la serie de líneas para los precios unitarios
fig.add_trace(go.Scatter(x=ventas_por_producto_enero['Producto'], y=precios_unitarios_enero, mode='lines', name='Precio Unitario', yaxis='y2'))

# Configurar el diseño del gráfico
fig.update_layout(
    title='Ventas Totales y Precios Unitarios por Producto en Enero',
    xaxis_title='Producto',
    yaxis_title='Ventas Totales',
    yaxis2=dict(
        title='Precio Unitario',
        overlaying='y',
        side='right'
    )
)

fig.show()
```
